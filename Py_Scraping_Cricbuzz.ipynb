```python
s_number = [2,9,6,13,4]
# few single contestess added manually 
#[number, price_pool, buy_in, no_of_spots,no_of_winners]
df_data=[['2','₹10 Crores','44','3000000','57'],
             ['2','₹1 Lakh','15','7777','51'],
             ['2','₹50000','179','333','38'],
             ['2','₹15 Lakh','99','20000','50'],
             ['2','₹5 Lakh','15','44333','36'],
             ['2','₹3 Lakh','75','5333','49']]





for number in s_number:
    driver = webdriver.Chrome()
    driver_history = []
    #open the site
#     number = 2
    site=f"https://www.dream11.com/cricket/contests/ipl/1027/14089?sectionIds[]={number}"
    # site="https://www.thirdlove.com/collections/underwear"
    driver_history.append(site)
    driver.get(site)
    driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)
    time.sleep(5)
    a_link = driver.find_elements_by_class_name("contestCardWrapper_0e13a")
    link_data=[]
    for a in a_link:
        price_pool         = a.find_element_by_class_name("prizePool_565d2").text
        buy_in             = a.find_element_by_class_name("currency-amount").text
        no_of_spots_left   = a.find_element_by_class_name("spotLefts_8d583").text
        no_of_spots        = a.find_element_by_class_name("totalSpots_b62ba").text
        no_of_winners       = a.find_element_by_class_name("iconLabelWrapper_43690").text
        print('price_pool: ',price_pool,'buy_in: ',buy_in,'no_of_spots_left: ',no_of_spots_left,'no_of_spots: ',no_of_spots,'no_of_winners: ',no_of_winners)
        link_data =[number, price_pool, buy_in, no_of_spots,no_of_winners]
        print('--------------------------------------------\n')
        df_data.append(link_data)    
    print('XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX')
    
    
    driver.close()
```

```python
import os
import glob
import pandas as pd
from pandas import ExcelFile
from pandas import ExcelWriter
 
def WriteDf(path,dataFrame,name,ind=True):
    #path='C:\\Users\\spai\\DB\\PO\\Results'
    extension = 'xlt'
    os.chdir(path)
    documents=os.listdir(path)
    if (name+'xlsx') in documents:
        os.remove(name+'xlsx')
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter(name+'.xlsx', engine='xlsxwriter')
 
    # Convert the dataframe to an XlsxWriter Excel object.
    dataFrame.to_excel(writer, sheet_name='Sheet1',index=ind,index_label='Index')
 
    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
#WriteDf(rawData,'Processed_data')
 

def WriteDf_Same_Sheet(path,dataFrameDict,name):
    #path='C:\\Users\\spai\\DB\\PO\\Results'
    extension = 'xlt'
    os.chdir(path)
    documents=os.listdir(path)
    if (name+'xlsx') in documents:
        os.remove(name+'xlsx')
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter(name+'.xlsx', engine='xlsxwriter')
    for key, value in dataFrameDict.items():
    # Convert the dataframe to an XlsxWriter Excel object.
        value.to_excel(writer, sheet_name=key[0:30],index=True,index_label='Index')
 
    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
#WriteDf(rawData,'Processed_data')
 
```
